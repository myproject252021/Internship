# 6. Dineout.co.in

```python
dineout=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
dineout
```

```python
soup1=BeautifulSoup(dineout.content)
soup1
```

```python
name=[]
for i in soup1.find_all('div',class_='restnt-info cursor'):
       name.append(i.text)

name
```

```python
cuisine=[]
for i in soup1.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.text.split()[6:8])

cuisine
```

```python
location=[]
for i in soup1.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.text)
    
location
```

```python
rating=[]
for i in soup1.findAll('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
    
rating   
```

```python
image=[]
for i in soup1.findAll('img',class_='no-img'):
    image.append(i['data-src'])
    
image    
```

```python
#Creating DataFrame
data1=pd.DataFrame({'Name':name,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Image URL':image})
data1
```
