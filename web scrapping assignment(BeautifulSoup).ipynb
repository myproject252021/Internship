# 1. Wikipedia.org

```python
#1) Write a python program to display all the header tags from wikipedia.org
web=requests.get('https://en.wikipedia.org/wiki/Main_Page')#taking permission from the website.
web
```

```python
content=BeautifulSoup(web.content)#accessing the content of the webpage.
content
```

```python
#to find the all the headers of wikipedia.org.

head=[] #making an empty list.
for i in content.find_all('span',class_='mw-headline'):
    head.append(i.text)
head
```

```python
#creating dataframe.
set=pd.DataFrame({'Header':head})
set
```

# 2.IMDB top 100 movies

```python
#2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame.
page=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')#taking permission from the webpage.
page
```

```python
imdb=BeautifulSoup(page.content)
imdb
```

```python
#extracting imdb's top 100 movies.
movies_name = [] #empty list
name = imdb.find_all("h3", class_="lister-item-header")
    
for i in name:
    
    for j in i.find_all("a"):
        movies_name.append(j.text)
    

movies_name
```

```python
#extracting imdb's top 100 movies rating.
rate=[]

for i in imdb.find_all('div',class_= "inline-block ratings-imdb-rating"):
    rate.append(i.text.replace('\n',''))

    

rate
```

```python
##extracting imdb's top 100 movies year of release.
year_of_release=[]
for i in imdb.find_all('span',class_='lister-item-year text-muted unbold'):
    year_of_release.append(i.text)

year_of_release
```

```python
Top_movie=pd.DataFrame({'Movies':movies_name,'Rating':rate,'Year of Release':year_of_release})
Top_movie
```

# 3.IMDB top 100 Indian Movies

```python
web=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
web
```

```python
soup2=BeautifulSoup(web.content)
soup2
```

```python
hindi_movie=[]

for i in soup2.find_all('td',class_='titleColumn'):
    hindi_movie.append(i.text.replace('\n',''))
    
    
hindi_movie
hindi_movie[:100]

```

```python
rate2=[]
for i in soup2.find_all('td',class_='ratingColumn imdbRating'):
    rate2.append(i.text.replace('\n',''))
    
rate2
rate2[:100]
```

```python
year_of_release2=[]
for i in soup2.find_all('span',class_='secondaryInfo'):
    year_of_release2.append(i.text.replace('\n',''))
year_of_release2
year_of_release2[:100]
    
```

```python
#creating DataFrame
indian_movie=pd.DataFrame({'Movies':hindi_movie,'Ratings':rate2,'Year_of_Release2':year_of_release2})
indian_movie
indian_movie[:100]
```

# 4.Meesho.com

```python
#product name, price and discounts
meesho=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
meesho
```

```python
product=BeautifulSoup(meesho.content)
product
```

```python
pc_name=[]
for i in product.find_all('p',class_='Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS'):
    
        pc_name.append(i.text.replace('\t',''))
pc_name
```

```python
price=[]
for i in product.find_all('h5',class_='Text__StyledText-sc-oo0kvp-0 hiHdyy'):
    price.append(i.text)

    
    
price
```

```python
discount=[]
for i in product.find_all('span',class_='Text__StyledText-sc-oo0kvp-0 lnonyH'):
    discount.append(i.text)
discount
```

```python
#Creating DataFrame
meesho=pd.DataFrame({'Product name':pc_name,'Price':price,'Discount':discount})
meesho
```

# 5.Bewakoof.com

```python
site1=requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts')
site1
```

```python
set1=BeautifulSoup(site1.content)
set1
```

```python
product_name=[]
for i in set1.find_all('div',class_="productCardDetail"):
    product_name.append(i.text)

product_name

```

```python
price=[]
for i in set1.find_all('span',class_='discountedPriceText'):
    price.append(i.text)

price
```

```python
image2=[]
for i in set1.find_all('img',class_='productImgTag'):
    image2.append(i['src'])
    
image2
```

```python
#creating dataFrame
bewakoof_1=pd.DataFrame({'Product Name':product_name,'Price':price,'Image URL':image2})
bewakoof_1
#There is only 10 product data itself.
```
